









# Preliminaries

## Imports

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import torch
import torch.nn as nn
from torch.utils.data import WeightedRandomSampler, DataLoader
# import datawig
import missingno as msno
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from collections import Counter
```

## Read Data

```python
data = pd.read_csv("dat.csv", index_col=0)
data.head()
```

```python
data.info(verbose=True)
```

# Data Preparation

## Prepare Data for Processing

Our interest lies in predicting readmission. For that reason, we can remove the patients that have died and remove the columns, `death.within.28.days`, `death.within.3.months` and `death.within.6.months`.

```python
re_data = data[(data["death.within.28.days"] != 1) & (data["death.within.3.months"] != 1) & (data["death.within.6.months"] != 1)]
re_data = re_data.drop(labels=['death.within.28.days', 'death.within.3.months', 'death.within.6.months'], axis='columns')
```

Additionally, we should merge our 3 target columns into one column and drop the remaining 3. This way, we'll have 0 representing no readmission, 1 representing readmission in 28 days, 2 representing readmission in 3 months, and 3 representing readmission in 6 months. After viewing the 3 target columns, this is under the assumption that:
1. If a patient is readmissed within 28 days, it means they've been readmissined within 3 and 6 months as well
2. If a patient is readmissed within 3 months, it means they've been readmissed within 6 months as well

```python
re_data['target'] = re_data['re.admission.within.28.days'] + re_data['re.admission.within.3.months'] + re_data['re.admission.within.6.months']
re_data = re_data.drop(labels=['re.admission.within.28.days', 're.admission.within.3.months', 're.admission.within.6.months'], axis='columns')
```

## Impute Missing Data

Now that our data is encoded, we can impute missing values. We will use a model to predict the missing values. We only want to consider columns that do not have too much data missing and we'll use a threshold of 50% for that. We'll also assume that the data is missing completely at random (MCAR) in order to perform this imputation.

```python
msno.bar(re_data)
```

```python
THRESHOLD = 0.5
drop_data = re_data.dropna(thresh=len(re_data) * THRESHOLD, axis=1)
msno.bar(drop_data)
```

```python
unique_counts = pd.DataFrame.from_records([(col, drop_data[col].nunique()) for col in drop_data.columns],
                          columns=['Column_Name', 'Num_Unique']).sort_values(by=['Num_Unique'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
unique_counts
```

From the data dictionary and the above unique counts dataframe, we should be able to set all the categories that have less than 10 unique values to categories. We can also drop `leukemia` as they are all the same value. Based on our knowledge as well, we know that the patient number is not relevant to prediction.

```python
drop_data = drop_data.drop(labels=['leukemia', 'inpatient.number'], axis='columns')
```

```python
for col in drop_data.columns:
    if drop_data[col].nunique() <= 10:
        drop_data[col] = drop_data[col].astype('category')
```

Now we can begin imputing by first encoding strings into numbers since `IterativeImputer` does not support strings.

```python
for col in drop_data.select_dtypes(['category']).columns:
    le = LabelEncoder()
    drop_data[col] = le.fit_transform(drop_data[col])
```

Now, we can use multiple imputation.

```python
imputer = IterativeImputer(max_iter=50, random_state=888)
imputer.fit(drop_data)
imputed = pd.DataFrame(imputer.transform(drop_data), columns=drop_data.columns)
msno.bar(imputed)
```

Since the imputer strips data types, we'll have to insert them back in.

```python
for col in imputed.columns:
    if imputed[col].nunique() < 10:
        imputed[col] = imputed[col].astype('category')
```

# Modelling

For convenience and GPU access, I'll export our imputed data and read it in if I had already had it saved here.

```python
# imputed.to_csv('imputed.csv')
# imputed = pd.read_csv('imputed.csv', index_col=0)
```

In order to prepare our data for modelling, we should use one hot encoding to encode our categorical data. We'll also create our train/test split.

```python
X, y = imputed.drop(labels=['target'], axis='columns'), imputed['target']
X = pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=888)
X_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

Let's also get an idea of the distribution of our target feature.

```python
freq = Counter(y)
freq
```

```python
plt.hist(y)
plt.show()
```

We have far more data that is classified into class 0 than the remaining classes. As a result, we should try to either use class weighting or oversample/undersample the underrepresented/overrepresented classes. Also, we'd typically would use a dimensionality reduction method such as PCA, but for Neural Networks, this can be implicitly done through the imput layer.

## Dataset

### Hyperparameters

```python
BATCH_SIZE = 256
EPOCHS = 300
LR = 0.0001
num_classes = 4
device = torch.device('cuda:0') if torch.cuda.is_available() else torch.device('cpu')
```

```python
class Heart(torch.utils.data.Dataset):
    def __init__(self, X, y):
        super().__init__()
        self.x = torch.tensor(X.values).to(device)
        self.t = torch.tensor(y.values).to(device)
        
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self, idx):
        return self.x[idx], self.t[idx]
```

```python
train_data = Heart(X_train, y_train)
test_data = Heart(X_test, y_test)
```

## Dataloaders

```python
cw_train_dl = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True)
test_dl = DataLoader(test_data, batch_size=BATCH_SIZE, shuffle=True)
```

Let's also create dataloaders trying oversampling. For sake of time, I'll use a set of weights for both methods to handle imbalanced data. I've set the baseline to be the ratios of class 0 to the remaining classes and we can tune the weights from that point on. Also, it would be more important to predict readmission accurately than say, no readmission, since it would be a more serious health concern.

```python
class_weights = [1., freq[0] / freq[1] + 1, freq[0] / freq[2] + 2, freq[0] / freq[3] + 0.75]
sample_weights = [0] * len(train_data)

for idx, (data, label) in enumerate(train_data):
    class_weight = class_weights[label.int()]
    sample_weights[idx] = class_weight
    
sampler = WeightedRandomSampler(sample_weights, num_samples=len(train_data), replacement=True)
s_train_dl = DataLoader(train_data, batch_size=BATCH_SIZE, sampler=sampler)
```

## Model Building

```python
class Network(nn.Module):
    def __init__(self, loss_fcn, input_dim, num_classes):
        super().__init__()
        self.losses = []
        self.test_losses = []
        self.input_dim = input_dim
        self.num_classes = num_classes
        self.loss_fcn = loss_fcn
        self.network = nn.Sequential(
            nn.Linear(input_dim, 60),
            nn.ReLU(),
            nn.Linear(60, 30),
            nn.ReLU(),
            nn.Linear(30, 15),
            nn.ReLU(),
            nn.Linear(15, 8),
            nn.ReLU(),
            nn.Dropout(),
            nn.Linear(8, num_classes),
#             nn.Softmax(dim=1) implementation in pytorch does not require this
        )
    
    def forward(self, x):
        return self.network(x)
    
    def learn(self, train_dl, test_dl, epochs=100, lr=0.01, plot=False):
        optim = torch.optim.SGD(self.parameters(), lr=lr)
        for epoch in tqdm(range(epochs)):
            total_loss = 0
            for x, t in train_dl:
                x = x.float().to(device)
                t = t.to(device)
                y = self(x)
                loss = self.loss_fcn(y, t.to(torch.int64))
                optim.zero_grad()
                loss.backward()
                optim.step()
                total_loss += loss.item()
            self.losses.append(total_loss / len(train_dl))
            
            test_loss = 0
            with torch.no_grad():
                for x, t in test_dl:
                    x = x.float().to(device)
                    t = t.to(device)
                    y = self(x)
                    loss = self.loss_fcn(y, t.to(torch.int64))
                    test_loss += loss.item()
                self.test_losses.append(test_loss / len(test_dl))
            
        if plot:
            plt.figure()
            plt.plot(self.losses, label='train')
            plt.plot(self.test_losses, label='test')
            plt.xlabel('Epochs')
            plt.ylabel('Loss')
            plt.legend()
    
    def test(self, dl):
        classes = [
            'no readmission',
            'readmission within 28 days',
            'readmission within 3 months',
            'readmission within 6 months'
        ]
        class_correct = [0, 0, 0, 0]
        class_total = [0, 0, 0, 0]
        y_true = []
        y_pred = []
        
        with torch.no_grad():
            for x, t in dl:
                x = x.float().to(device)
                t = t.to(device)
                outputs = self(x)
                pred = torch.argmax(outputs, dim=1)
                y_true.extend(t.cpu().numpy())
                y_pred.extend(pred.cpu().numpy())
                for i in range(len(pred)):
                    if t[i] == pred[i]:
                        class_correct[t[i].int()] += 1
                    class_total[t[i].int()] += 1
        
        for i in range(4):
            print(f'Accuracy of {classes[i]}: {100 * class_correct[i] / class_total[i]}')
        
        cm = confusion_matrix(y_true, y_pred)
        display = ConfusionMatrixDisplay(cm)
        display.plot()
```

## Training

First we try training a model where we use class weighting.

```python
loss_fcn = nn.CrossEntropyLoss(weight=torch.tensor(class_weights))
netA = Network(loss_fcn, 115, num_classes)
netA.to(device)
netA.learn(cw_train_dl, test_dl, epochs=EPOCHS, plot=True)
```

Then we try training a model where we use weighted sampling.

```python
netB = Network(loss_fcn, 115, num_classes)
netB.to(device)
netB.learn(s_train_dl, test_dl, epochs=EPOCHS, plot=True)
```

## Evaluation

Results for the first network.

```python
netA.test(test_dl)
```

Results for the second network.

```python
netB.test(test_dl)
```

## Aside

It might just be the case that a neural network is not appropriate for this type of data. Let us just try a base XGBoost model.

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(X_train, y_train)
prediction = xgb.predict(X_test)

cm = confusion_matrix(y_test, prediction)
display = ConfusionMatrixDisplay(cm)
display.plot()
```

Compared to the neural networks, it performed quite well. With proper tuning, it may even perform better, although this could also be said for the neural network models.
